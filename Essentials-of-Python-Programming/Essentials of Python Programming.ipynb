```python
import random
```

```python
while True:
    
    # Display the choices to the user
    print("\nEnter 1 if you would like to play Rock, Paper, Scissors")
    print("Enter 2 if you would like to play Rock, Paper, Scissors, Lizard, Spock\n")
    
    # Accept the users choice using input() and store it in a variable after converting it into an integer
    game_choice=input()
    
    # Checking whether the user has entered a valid digit. 
    # If the user has not entered a valid digit, then start a new iteration of the loop using "continue" so that the user can enter a valid choice
    # If the user has entered a valid digit, we convert the variable from string to integer type
    if(game_choice.isdigit() == False):
        print("You have not entered a valid choice")
        continue
    else:
        game_choice=int(game_choice)
        
    # Checking whether the user has entered a valid choice (1 or 2)
    # If the user has not entered a valid choice, then start a new iteration of the loop using "continue" so that the user can enter a valid choice
    if(game_choice not in [1,2]):
        print("You have not entered a valid choice")
        continue
        
    # If the user has entered "1", then proceed with the game "Rock, Paper, and Scissors"
    if(game_choice==1):
        
        # Accepting the users action in the game and storing it in a variable
        player1_action= input("Enter your choice (rock, paper, or scissors)\n")
        
        # Converting the users action to lower case to avoid case sensitivity
        player1_action=player1_action.lower()
        
        # Storing all of the accepted actions in a list
        actions_list = ["rock","paper","scissors"]
        
        # Using the random module to choose a random integer between 0 and 2 (inclusive)
        # The computers action in the game is determined by accessing the value at the random integer's index in actions_list
        player2_action=actions_list[random.randint(0,2)]
        
        # We display both players actions in the game
        print("\nPlayer 1 (You): ", player1_action," Player 2 (Computer): ", player2_action)
        
        # Checking if both players chose the same action
        # If so we print that the result is a tie
        if(player1_action==player2_action):
            print("Its a tie!")
            
        # Similarly we use nested conditional statements to check each player's actions and display the results of the game
        elif(player2_action=="paper"):
            if(player1_action=="rock"):
                print("Paper covers rock. You lose")
            if(player1_action=="scissors"):
                print("Scissors cuts paper. You are the winner!")
            
        elif(player2_action=="scissors"):
            if(player1_action=="rock"):
                print("Rock crushes scissors. You are the winner!")
            if(player1_action=="paper"):
                print("Scissors cuts paper. You lose")
            
        elif(player2_action=="rock"):
            if(player1_action=="scissors"):
                print("Rock crushes scissors. You lose")
            if(player1_action=="paper"):
                print("Paper covers rock. You are the winner!")
            
    # If the user entered "2", then proceed with the game "Rock, Paper, Scissors, Lizard, and Spock"        
    elif(game_choice==2):
        
        # Accepting the users action in the game and storing it in a variable
        player1_action= input("Enter your choice (rock, paper, scissors, lizard, spock)\n")
        
        # Converting the users action to lowercase to avoid case sensitivity
        player1_action=player1_action.lower()
        
        # Storing all of the accepted actions in a list
        actions_list = ["rock","paper","scissors","lizard","spock"]
        
        # Using the random module to choose a random integer between 0 and 4 (inclusive)
        # The computers action in the game is determined by accessing the value at the random integer's index in actions_list
        player2_action=actions_list[random.randint(0,4)]
        
        # We display both players actions in the game
        print("\nPlayer 1 (You): ", player1_action," Player 2 (Computer): ", player2_action)
        
        # Checking if both players chose the same action
        # If so we print that the result is a tie
        if(player1_action==player2_action):
            print("Its a tie!")
            
        # Similarly we use nested conditional statements to check each player's actions and display the results of the game    
        elif(player2_action=="paper"):
            if(player1_action=="rock"):
                print("Paper covers rock. You lose")
            if(player1_action=="scissors"):
                print("Scissors cuts paper. You are the winner!")
            if(player1_action=="lizard"):
                print("Lizard eats paper. You are the winner!")
            if(player1_action=="spock"):
                print("Paper disproves spock. You lose")
            
        elif(player2_action=="scissors"):
            if(player1_action=="rock"):
                print("Rock crushes scissors. You are the winner!")
            if(player1_action=="paper"):
                print("Scissors cuts paper. You lose")
            if(player1_action=="lizard"):
                print("Scissors decapitates lizard. You lose")
            if(player1_action=="spock"):
                print("Spock smashes scissors. You are the winner!")
            
        elif(player2_action=="rock"):
            if(player1_action=="scissors"):
                print("Rock crushes scissors. You lose")
            if(player1_action=="paper"):
                print("Paper covers rock. You are the winner!")
            if(player1_action=="lizard"):
                print("Rock crushes lizard. You lose")
            if(player1_action=="spock"):
                print("Spock vaporizes rock. You are the winner!")
        
        elif(player2_action=="lizard"):
            if(player1_action=="scissors"):
                print("Scissors decapitates lizard. You are the winner!")
            if(player1_action=="paper"):
                print("Lizard eats paper. You lose")
            if(player1_action=="rock"):
                print("Rock crushes lizard. You are the winner!")
            if(player1_action=="spock"):
                print("Lizard poisons spock. You lose")
        
        elif(player2_action=="spock"):
            if(player1_action=="scissors"):
                print("Spock smashes scissors. You lose")
            if(player1_action=="paper"):
                print("Paper disproves spock. You are the winner!")
            if(player1_action=="rock"):
                print("Spock vaporizes rock. You lose")
            if(player1_action=="lizard"):
                print("Lizard poisons spock. You are the winner!")
    
    # Asking the user whether they would like to play another round
    continue_game_choice=input("\nWould you like to play another round? (Enter 'Q' to exit, Enter 'Y' to keep playing)\n")
    
    # Converting the users answer to uppercase to avoid case sensitivity
    continue_game_choice=continue_game_choice.upper()
    
    # If the user enters "Y" then the next iteration of the loop starts so the user can play another round of the game
    # Whereas if the user enters "Q" then "break" is used to exit the loop and end the game
    if(continue_game_choice=="Y"):
        pass
    elif(continue_game_choice=="Q"):
        print("Thanks for playing!")
        break
    
    
```
